## 3. Matrice de corrélation entre les variables numériques

```python
# Variables vides dans la heatmap
col_to_drop = ['MMS', 'r', 'De', 'Vf', 'year', 'Ernedc (g/km)']

# Variables numériques, avec le target en première position
df_num = df.select_dtypes('number').drop(columns=col_to_drop)
df_num = df_num[['Ewltp (g/km)'] \
                + [col for col in df_num.columns if col != 'Ewltp (g/km)']]

plt.figure(figsize=(15, 8))
sns.heatmap(df_num.corr(), annot = True, center = 0, cmap = "RdBu_r");
plt.title("Matrice de corrélation entre les variables numériques."
          " La cible est à gauche.\n",
          fontweight="bold");
```

```python
"""
Suppression des variables inutiles (trop de catégories)
"""
df = df.drop(columns=['Mh', 'Cr', 'Ct', 'Cn', 'Mk'])
df.info()
```

```python
"""
Suppression des variables inutiles (non pertinentes)
"""
df = df.drop(axis=1, columns=['Country', 'VFN', 'Va', 'Ve', 'Date of registration'])
```

```python
"""
Suppression des variables inutiles - suite
"""
df = df.drop(axis=1, columns=['Man', 'Tan', 'T', 'IT'])
```

## 2. Gestions des NaNs

```python
df.isna().sum()
```

```python
# remplacement à 0 des lignes Erwltp, ec, ep, elec range, z (Wh/km) avec des NA
df[["Erwltp (g/km)", "ec (cm3)", "ep (KW)", "Electric range (km)", "z (Wh/km)"]] = df[["Erwltp (g/km)", "ec (cm3)", "ep (KW)", "Electric range (km)", "z (Wh/km)"]].fillna(0)
```

```python
# Suppresion des lignes Fuel consumption avec des NA
df= df.dropna(subset="Fuel consumption ")

# suppression des lignes Ewltp avec des NA
df = df.dropna(subset=["Ewltp (g/km)"])

# remplacement par la moyenne des lignes W, At1 avec des NA
df["W (mm)"] = df["W (mm)"].fillna(df["W (mm)"].mean())
df["At1 (mm)"] = df["At1 (mm)"].fillna(df["At1 (mm)"].mean())

# suppression des lignes m (kg) avec des NA
df = df.dropna(subset=["m (kg)"])
```

```python
# Remplacement des NaNs de Mp par le mode
df['Mp'] = df['Mp'].fillna(df['Mp'].mode()[0])
```

```python
df.isna().sum()
```

## 3. Encodage des variables catégorielles

### a) OneHotEncoding pour les variables n'ayant pas trop de catégories

```python
df = pd.get_dummies(df, columns=['Ft', 'Fm'])
df.info()
```

### b) Frequency Encoder pour Mp (trop de catégories)

```python
frequency = df['Mp'].value_counts(normalize=True)

df['Mp'] = df['Mp'].map(frequency)
```
